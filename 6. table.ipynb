## 조달 사업 상장/비상장 비교 테이블(표2)

```python
df_bid = pd.read_parquet(path+ 'orign_bid.pqt', engine = 'pyarrow') # bid 원본 데이터
df_bid
```

```python
df_bid['공고게시일자'] = (df_bid['공고게시일자'].values / 1e2).astype(int) # year 컬럼 생성
df_bid['연도'] = (df_bid['공고게시일자'].values / 1e2).astype(int) # year 컬럼 생성
df_bid.drop(columns = '공고게시일자', inplace = True)
df_bid = df_bid.reindex(columns = ['연도','입찰공고번호', '입찰업체사업자등록번호', '수요기관구분', '입찰금액', '낙찰여부'])
df_bid # 사업자 등록번호 int임
```

```python
## 교육기관이 조달금액의 규모가 가장 작음으로 제외
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '지방공기업', '공기업', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '준정부기관', '국가기관', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '기타공공기관', '국가기관', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '지자체 출자출연기관', '지자체', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '기타기관', '국가기관', df_bid.수요기관구분)
df_bid.drop(df_bid.loc[df_bid['수요기관구분'] == '교육기관','수요기관구분'].index, inplace = True)
df_bid['수요기관구분'].value_counts()
```

```python
df_Y = df_bid.loc[df_bid.낙찰여부 == 'Y', :]
df_Y.dropna(inplace = True)
df_Y.rename(columns = {'입찰업체사업자등록번호': '사업자등록번호'}, inplace = True)
df_Y # 낙찰 받은 애들 데이터
```

```python
tobinsQ_tt= pd.read_parquet(path + 'tobinsQ_final.pqt', engine = 'pyarrow') # tobinsQ 정리 데이터
tobinsQ_tt0=tobinsQ_tt.groupby(['entity_name','연도', '사업자등록번호']).agg({'market':'first'})
tobinsQ_tt0.reset_index(inplace = True)
tobinsQ_tt0
```

```python
col = ['연도', '사업자등록번호']
tobinsQ_Y = pd.merge(df_Y, tobinsQ_tt0, how = 'left', on = col) # bid 데이터를 기준으로 tobinsQ 묶은 데이터 -> 상장/비상장 기업 모두 있는 데이터
tobinsQ_Y
```

```python
tobinsQ_Y.입찰공고번호.nunique()
```

```python
lst_company = tobinsQ_Y.loc[tobinsQ_Y['entity_name'].notnull(), :] # 상장기업
unlst_company = tobinsQ_Y.loc[tobinsQ_Y['entity_name'].isnull(), :] # 비상장기업
display(lst_company.head(), unlst_company.head())
## 예시 lst_company 4,5번 행 엑셀상에서도 대한 과학은 같은 공고에 3번 받음 -> 입찰금액 sum()
```

## 시장별 입찰 및 낙찰 여부 (표4)

```python
total_df = pd.read_parquet(path +'total_df.pqt', engine = 'pyarrow')
total_df.groupby(['시장', '입찰여부','낙찰여부']).agg({'낙찰여부': np.size, '매출':np.mean, '영업이익률': np.mean, '연구개발비':np.mean}).sort_values(by='시장', ascending = False)
```

```python
# 마지막으로 본데이터를 활용하여 낙찰 횟수를 비교해서(낙찰횟수로 그룹을 나누어) 시장별 재무, 토빈스Q 비교(표5) 
total_dfm = total_df[total_df['입찰여부'] == True]
total_dfm.groupby(['시장','낙찰여부']).agg({'낙찰여부': np.size, '입찰횟수_공기업':np.mean, '입찰횟수_지자체': np.mean, '입찰횟수_국가기관':np.mean, '낙찰수_공기업': np.mean, '낙찰수_지자체': np.mean, '낙찰수_국가기관':np.mean })
```

```python
n = len(total_dfm)
dt = (total_dfm.groupby(['시장']).size() / n) * 100
dt.sort_values(ascending = False)
```

```python
# Kospi 입찰기업 중 상위 TOP3 업종별 낙찰비율 (표6)
kospi = total_df[total_df['시장'] == 'KOSPI']
kospi_t = kospi[kospi['입찰여부'] == True] 
kospi_t.loc[~kospi_t['업종분류코드'].isin(['C', 'H', 'J']), '업종분류코드'] = 'etc'
kospi_t.groupby(['업종분류코드', '낙찰여부']).agg({'낙찰여부': np.size, '입찰횟수_공기업':np.mean, '입찰횟수_지자체': np.mean, '입찰횟수_국가기관':np.mean, '낙찰수_공기업': np.mean, '낙찰수_지자체': np.mean, '낙찰수_국가기관':np.mean})
```

```python
n = len(kospi)
dt = (kospi.groupby(['업종분류코드']).size() / n) * 100
dt.sort_values(ascending = False)
```

```python
# Kosdaq 입찰기업 중 상위 TOP3 업종별 낙찰비율(표7)
kosdaq = total_df[total_df['시장'] == 'KOSDAQ']
kosdaq_t = kosdaq[kosdaq['입찰여부'] == True] 
kosdaq_t.loc[~kosdaq_t['업종분류코드'].isin(['C', 'J', 'G']), '업종분류코드'] = 'etc'
kosdaq_t.groupby(['업종분류코드', '낙찰여부']).agg({'낙찰여부': np.size, '입찰횟수_공기업':np.mean, '입찰횟수_지자체': np.mean, '입찰횟수_국가기관':np.mean, '낙찰수_공기업': np.mean, '낙찰수_지자체': np.mean, '낙찰수_국가기관':np.mean})
```

```python
n = len(kosdaq)
dt = (kosdaq.groupby(['업종분류코드']).size() / n) * 100
dt.sort_values(ascending = False)
```
