```python
col = ['symbol', '연도', '분기']
df_Tobin = pd.merge(fin_tt, market_tt, how = 'left', on = col) # 재무정보 데이터에 맞게 left join
df_Tobin
```

```python
df_Tobin.drop(df_Tobin[df_Tobin.시가총액.isnull()].index, axis = 0, inplace = True) # 시가총액 없는 애들 다 삭제 
df_Tobin['TobinsQ'] = (df_Tobin['시가총액'] + df_Tobin['부채'] - df_Tobin['유동자산']) / df_Tobin['자산'] ## 토빈스Q 계산 //  '-'가 나오면 안된다.
indx=df_Tobin[df_Tobin.TobinsQ >= float('inf')].index # inf 나온 index 추출
df_Tobin.drop(index = indx, axis = 0, inplace = True) # 자산, 유동자산, 부채가 없는 애들 dropna
df_Tobin.drop(columns = ['자산', '유동자산', '부채', '시가총액'], axis = 1, inplace = True)
df_Tobin.sort_values(by=['entity_name', '연도', '분기'], inplace = True)
df_Tobin
```

## 종업원수 데이터 merge

```python
df_tobinsQ['symbol'] = df_tobinsQ['symbol'].apply(lambda x: x[4:]) # 차후 데이터 병합을 위한 primary key 정리 
df_tobinsQ['symbol'] = df_tobinsQ['symbol'].astype(float) # symbol type 정리
```

```python
from glob import glob
path = 'C:/Users/0630r/inha2/Research/Dart/Dart_file/support_data/employee/'

list_20 = sorted(glob(path + "/2020/*.xls"))
list_21 = sorted(glob(path + "/2021/*.xls"))
list_22 = sorted(glob(path + "/2022/*.xls"))  
#folders= os.listdir('marcap/data')
#for files in forders:
lst = np.array([list_20, list_21, list_22])
```

```python
dfs = {} # 데이터 프레임 저장

lst = np.array([list_20, list_21, list_22])

for year_files in lst:
    year = int(year_files[0][-10:-6])
    df_year = pd.DataFrame()
    n = 1

    for file in year_files:
        df = pd.read_excel(file, header=2)
        df.insert(3, '분기', str(n) + 'Q') # 분기추가
        df.insert(2, '연도', year) # 연도추가
        df['종목코드']  = df['종목코드'].astype(float)
        df_year = pd.concat([df_year, df], axis=0)
        df_year = df_year[['회사명', '종목코드','연도', '분기', '사업부문', '성별', '합계']]
        n += 1

    key_name = 'df_' + str(year)
    dfs[key_name] = df_year

df_2020 = dfs['df_2020']
df_2021 = dfs['df_2021']
df_2022 = dfs['df_2022']
```

```python
df_employ = pd.concat([df_2020, df_2021, df_2022], axis = 0)
df_employ.info()
```

```python
df_employ.head()
```

```python
df_employ[df_employ['합계'] == '-']
```

```python
df_employ.dropna(axis = 0, inplace = True) # null값 인덱스별 삭제
df_employ['합계'].replace('-', 0, inplace = True) # -를 0으로 바꾸기
df_employ = df_employ.astype({'합계': 'int'}) # 합계를 int로 바꾸기
df_employ = df_employ.groupby(['회사명','연도', '분기'], as_index = False).agg({'종목코드': 'mean',
                                                                              '합계': 'sum'}) # 회사명으로 groupby로 바꾸기
df_employ.rename(columns = {'종목코드': 'symbol'}, inplace = True)
df_employ
```

```python
col = ['symbol', '연도', '분기']
tobinsQ_tt = pd.merge(df_tobinsQ, df_employ, on = col)
tobinsQ_tt.drop(columns = ['종업원수', '회사명'], axis =1, inplace = True)
tobinsQ_tt.rename(columns = {'합계': '종업원수'}, inplace = True)
```

```python
tobinsQ_tt[tobinsQ_tt['종업원수'] == 0] # 69개의 기업 / 178개의 행이 종업원 수 기록 X 
```

```python
idx = tobinsQ_tt[tobinsQ_tt['종업원수'] == 0].index ## 종업원 수가 공지되지 않은 기업 전부 삭제
tobinsQ_tt.drop(idx, inplace = True)
tobinsQ_tt.reset_index(drop=True, inplace = True)
```
