## 2020년 Market 정리

```python
df2 = df2[3]

def start(df):
    df = df.transpose() # 행열 변환
    df.drop([0,1], axis = 1, inplace = True) # 필요없는 열 삭제
    df = df.rename(columns = df.iloc[0,:]) # 1행을 열로 올림
    df = df.drop(df.index[0]) # 원래 1행 삭제
    df.columns = df.columns.astype('string')
    return df

df2 = start(df2)
df2.to_parquet('orign_market20.pqt') # 20년도 저장
df2.head()
```

```python
path = 'C:/Users/0630r/inha2/Research/Dart/Dart_file/real_code/pqt_file/'
df2 = pd.read_parquet(path+'orign_market20.pqt')
df2.head()
```

```python
# 분기별 데이터 subset
df20_1Q = df2.loc[:, :'2020-03-31']
df20_2Q = df2.loc[:, '2020-04-01':'2020-06-30']
df20_3Q = df2.loc[:, '2020-07-01':'2020-09-30']
df20_4Q = df2.loc[:, '2020-10-01':'2020-12-31']
```

```python
df20_1Q.head(6)
# (시가총액 , 종가) | (시가총액,  종가) 로 이루어져있음
```

```python
# 시가총액, 주가 평균 정리
def Mean(df):
    df_시총 = df[::2] # 시가총액 추출
    df_시총['시가총액'] = round(df_시총[df.columns].sum(axis = 1)/len(df.columns)) # 시가 총액 평균 
    df_주가 = df[1::2] # 주가 추출
    df_주가['주가평균'] = round(df_주가[df.columns].sum(axis = 1)/len(df.columns)) # 종가 평균
    df_주가.set_index(df_시총.index, inplace = True) # 인덱스를 기업 이름으로 정렬
    df1 = pd.concat([df_시총,df_주가], axis = 1) # concat
    df1 = df1[['시가총액', '주가평균']]
    return df1

# 데이터 합병 및 정리
def clean(df1,df2,df3,df4):
    df20 = pd.concat([df1, df2, df3, df4], axis = 0) # 전체 concat
    df20 = df20[['분기', '시가총액', '주가평균']] # 
    df20.reset_index(inplace = True) # 인덱스 추가
    df20.rename(columns = {'index': '회사명'}, inplace = True) # 회사명 열 생성
    df20.sort_values(by=['회사명','분기'], inplace = True) # 회사와 분기별로 정렬
    df20.reset_index(drop = True, inplace = True)
    return df20 
```

```python
df1Q = Mean(df20_1Q)
df1Q['분기'] = '1Q'
df2Q = Mean(df20_2Q)
df2Q['분기'] = '2Q'
df3Q = Mean(df20_3Q)
df3Q['분기'] = '3Q'
df4Q = Mean(df20_4Q)
df4Q['분기'] = '4Q'

df20 = clean(df1Q, df2Q, df3Q, df4Q)
df20.insert(1,'연도', 2020) # 2020 년도 추가
df20.drop(df20[df20['시가총액'] == 0].index, inplace = True) # 시가총액 0인 분기 삭제 
df20
```
