## 조달에 참여한 기업 / 조달에 참여하지 않은 기업 -> 즉 TobinsQ 데이터에 조달 데이터 left join

```python
import pandas as pd
from glob import glob
import numpy as np
import copy
path = 'C:/Users/0630r/inha2/Research/Dart/Dart_file/support_data'
```

```python
list_20 = sorted(glob(path + "/2020/*.csv"))
list_21 = sorted(glob(path + "/2021/*.csv"))
list_22 = sorted(glob(path + "/2022/*.csv"))  
#folders= os.listdir('marcap/data')
#for files in forders:
```

```python
df_2020 = pd.DataFrame()
for file in list_20:
    df = pd.read_csv(file)
    df_2020 = pd.concat([df_2020, df], axis = 0)
    df_2020 = df_2020[['등록유형', '공고게시일자','입찰공고번호','공고기관명', '입찰업체사업자등록번호', '수요기관구분', '참가수', '입찰금액', '낙찰여부', '업종제한여부', '업종제한내용', '지역제한내용']]

df_2021 = pd.DataFrame()
for file in list_21:
    df = pd.read_csv(file)
    df_2021 = pd.concat([df_2021, df], axis = 0)
    df_2021 = df_2021[['등록유형', '공고게시일자','입찰공고번호', '공고기관명', '입찰업체사업자등록번호', '수요기관구분', '참가수', '입찰금액', '낙찰여부', '업종제한여부', '업종제한내용', '지역제한내용']]

df_2022 = pd.DataFrame()
for file in list_22:
    df = pd.read_csv(file)
    df_2022 = pd.concat([df_2022, df], axis = 0)
    df_2022 = df_2022[['등록유형', '공고게시일자','입찰공고번호', '공고기관명', '입찰업체사업자등록번호', '수요기관구분', '참가수', '입찰금액', '낙찰여부', '업종제한여부', '업종제한내용', '지역제한내용']]

```

```python
# 사업자등록번호 , 연도, 입찰횟수, 낙찰금액, 경쟁횟수
df_bid = pd.concat([df_2020, df_2021, df_2022], axis = 0)
df_bid.reset_index(drop=True, inplace = True)
df_bid.drop(df_bid[df_bid['입찰업체사업자등록번호'] == 'F000000696'].index, inplace = True)
df_bid.drop(df_bid[df_bid['입찰업체사업자등록번호'] == 'F000000809'].index, inplace = True) # 폴란드의 금속 가공 회사 -> 주가 데이터에 존재하지 않아서 삭제
df_bid['입찰업체사업자등록번호'] = pd.to_numeric(df_bid['입찰업체사업자등록번호'])
```

```python
import re
import numpy as np
 
df_bid['참가수'] = df_bid['참가수'].apply(lambda x: re.sub(',', '', str(x)) if re.findall('\d+', str(x)) else np.nan) # type => int로 변경
df_bid['참가수'] = pd.to_numeric(df_bid['참가수']) 
df_bid['입찰금액'] = df_bid['입찰금액'].apply(lambda x: re.sub(',', '', str(x)) if re.findall('\d+', str(x)) else np.nan) # type => int로 변경
df_bid['입찰금액'] = pd.to_numeric(df_bid['입찰금액'])
df_bid['입찰공고번호'] = df_bid['입찰공고번호'].apply(lambda x: re.sub('[,-]', '', str(x)) if re.findall('\d+', str(x)) else np.nan) # [, -] 제거
```

```python
df_bid.to_parquet('orign_bid.pqt') ## 원본 bid 데이터 저장
```

```python
path = 'C:/Users/0630r/inha2/Research/Dart/Dart_file/real_code/pqt_file/'
df_bid = pd.read_parquet(path + 'orign_bid.pqt', engine = 'pyarrow')
df_bid
```

```python
df_bid['공고게시일자'] = (df_bid['공고게시일자'].values / 1e2).astype(int) # year 컬럼 생성
df_bid['연도'] = (df_bid['공고게시일자'].values / 1e2).astype(int) # year 컬럼 생성
df_bid['분기'] = df_bid['공고게시일자'].apply(lambda x: int(str(x)[4:])) # 분기 생성
df_bid['분기'] = (df_bid['분기'] - 1) // 3 + 1 
df_bid['분기'] = df_bid['분기'].apply(lambda x: str(x)+'Q')
df_bid.drop(columns = '공고게시일자', inplace = True)
df_bid = df_bid.reindex(columns = ['연도', '분기', '입찰업체사업자등록번호', '수요기관구분', '참가수', '입찰금액', '낙찰여부'])
df_bid.dropna(inplace = True)
df_bid.rename(columns = {'입찰업체사업자등록번호':'사업자등록번호'}, inplace = True)
df_bid
```

### 1. bid_nodemand data

```python
## 기업, 연도, 분기, 기관별 1.낙찰수, 2. 입찰금액sum, 3. 입찰 참여 수 4. 낙찰사업 참여자수 
## 입찰 횟수
df_size = df_bid.groupby(['연도','분기', '사업자등록번호']).size() # 사업자등록번호별 입찰횟수
df_size.name = '입찰횟수'
df_size = pd.DataFrame(df_size)
df_size = df_size.reset_index()
df_size
```

```python
## 낙찰횟수, 낙찰사업 평균참여수, 낙찰금액
df_Y = df_bid.loc[df_bid.낙찰여부 == 'Y', :] # 낙찰한 기업들 
df_Y = df_Y.groupby(['연도', '분기', '사업자등록번호']).agg({'참가수': np.mean, '입찰금액': np.sum, '낙찰여부': np.size})
df_Y.reset_index(inplace = True)
df_Y.rename(columns = {'참가수': '낙찰사업평균참여수', '낙찰여부': '낙찰수', '입찰금액': '낙찰금액'}, inplace = True)
df_Y
```

```python
col = ['연도', '분기', '사업자등록번호']
bid_tt = pd.merge(df_size, df_Y, how = 'left', on = col) # 입찰횟수 data + 낙찰횟수 data
bid_tt.fillna(0, inplace = True)
bid_tt
```

### 2. bid_final data

```python
# 낙찰 Y인 애들은 기준 평균 
bid_Y = df_bid[df_bid['낙찰여부'] == 'Y']
bid_Y.groupby('수요기관구분')['입찰금액'].mean().sort_values()
```

```python
# Top 3 기관 
# 공기업(공기업, 지방공기업)
# 정부기관(준정부기관, 국가기관, 기타공공기관, 기타기관)
# 지자체(지자체 출자출연기관, 지자체)
# 교육기관 삭제
```

```python
## 교육기관이 조달금액의 규모가 가장 작음으로 제외
## 수요이관 변환
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '지방공기업', '공기업', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '준정부기관', '국가기관', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '기타공공기관', '국가기관', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '지자체 출자출연기관', '지자체', df_bid.수요기관구분)
df_bid['수요기관구분'] = np.where(df_bid['수요기관구분'] == '기타기관', '국가기관', df_bid.수요기관구분)
df_bid.drop(df_bid.loc[df_bid['수요기관구분'] == '교육기관','수요기관구분'].index, inplace = True)
df_bid['수요기관구분'].value_counts()
```

```python
## 기업, 연도, 분기, 기관별 1.낙찰수, 2. 입찰금액sum, 3. 입찰 참여 수 4. 낙찰사업 참여자수 
```

```python
## 입찰 횟수
df_size1 = df_bid.groupby(['연도','분기', '사업자등록번호', '수요기관구분']).size()
df_size1.name = '입찰횟수'
df_size1 = pd.DataFrame(df_size1)
df_size1 = df_size1.reset_index()
df_size1
```

```python
## 낙찰횟수, 낙찰사업 평균참여수, 낙찰금액
df_Y1 = df_bid.loc[df_bid.낙찰여부 == 'Y', :]
df_Y1 = df_Y1.groupby(['연도', '분기', '사업자등록번호', '수요기관구분']).agg({'참가수': np.mean, '입찰금액': np.sum, '낙찰여부': np.size})
df_Y1.reset_index(inplace = True)
df_Y1.rename(columns = {'참가수': '평균참여수', '낙찰여부': '낙찰수', '입찰금액': '낙찰금액'}, inplace = True)
df_Y1
```

```python
# 연도별 분기별 기업별 수요기관별 
col = ['연도', '분기', '사업자등록번호', '수요기관구분']
bid_tt1 = pd.merge(df_size1, df_Y1, how = 'left', on = col)
bid_tt1.fillna(0, inplace = True)
bid_tt1
```

```python
## 여기서부터 수요기관별 열 추가하여 정리
bid_tt1['입찰횟수(공기업)'] = np.where(bid_tt1.수요기관구분 == '공기업', bid_tt1.입찰횟수, 0) # np.where(조건, T, F)
bid_tt1['평균참여수(공기업)'] = np.where(bid_tt1['수요기관구분'] == '공기업', bid_tt1['평균참여수'], 0)
bid_tt1['낙찰금액(공기업)'] = np.where(bid_tt1.수요기관구분 == '공기업', bid_tt1.낙찰금액, 0)
bid_tt1['낙찰수(공기업)'] = np.where(bid_tt1.수요기관구분 == '공기업', bid_tt1.낙찰수, 0)

bid_tt1['입찰횟수(지자체)'] = np.where(bid_tt1.수요기관구분 == '지자체', bid_tt1.입찰횟수, 0)
bid_tt1['평균참여수(지자체)'] = np.where(bid_tt1['수요기관구분'] == '지자체', bid_tt1['평균참여수'], 0)
bid_tt1['낙찰금액(지자체)'] = np.where(bid_tt1.수요기관구분 == '지자체', bid_tt1.낙찰금액, 0)
bid_tt1['낙찰수(지자체)'] = np.where(bid_tt1.수요기관구분 == '지자체', bid_tt1.낙찰수, 0)

bid_tt1['입찰횟수(국가기관)'] = np.where(bid_tt1.수요기관구분 == '국가기관', bid_tt1.입찰횟수, 0)
bid_tt1['평균참여수(국가기관)'] = np.where(bid_tt1['수요기관구분'] == '국가기관', bid_tt1['평균참여수'], 0)
bid_tt1['낙찰금액(국가기관)'] = np.where(bid_tt1.수요기관구분 == '국가기관', bid_tt1.낙찰금액, 0)
bid_tt1['낙찰수(국가기관)'] = np.where(bid_tt1.수요기관구분 == '국가기관', bid_tt1.낙찰수, 0)
```

```python
bid_tt1 = bid_tt1.groupby(['연도', '분기', '사업자등록번호']).sum() 
bid_tt1.reset_index(inplace = True)
bid_tt1.drop(columns = ['입찰횟수', '평균참여수', '낙찰수'], axis = 1, inplace = True)
bid_tt1
```

```python
bid_tt1.to_parquet('bid_final.pqt')
# 완
```
